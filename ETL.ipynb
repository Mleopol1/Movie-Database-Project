```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import openpyxl
```

```python
# Import csv file and convert to pandas dataframe
csv_file = "netflix_titles.csv"
netflix_df = pd.read_csv(csv_file)
netflix_df.head()
```

```python
# Check all of the column names
netflix_df.columns
```

```python
# Drop unneccesary columns
netflix_clean = netflix_df.drop(columns=['show_id', 'director', 'cast', 'country',\
                                         'listed_in', 'description', 'duration']).copy()
```

```python
netflix_clean.head()
```

```python
# Import json file and convert to pandas dataframe
json_file = "movie.json"
imdb_df = pd.read_json(json_file)
imdb_df.head()
```

```python
# Check all of the column names
imdb_df.columns
```

```python
# Remove unnecessary columns
imdb_clean = imdb_df.drop(columns=['rating', 'year', 'img_url', 'countries', 'languages',\
                                   'actors', 'genre', 'tagline', 'description', 'directors',\
                                  'runtime']).copy()
```

```python
# Remove commas from votes column to allow values to be int
imdb_clean['votes'] = imdb_clean['votes'].str.replace(',', '')
```

```python
imdb_clean.head()
```

```python
# Connect to local db
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'im_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
# check to make sure table names exist
engine.table_names()
```

```python
# Load csv-converted dataframe into to SQL
netflix_clean.to_sql(name='netflix', con=engine, if_exists='append', index=False)
```

```python
# Load json-converted dataframe into to SQL
imdb_clean.to_sql(name='imdb', con=engine, if_exists='append', index=False)
```
